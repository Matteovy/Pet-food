## Plotting Cap Curve

#### Capping Value analysis of "freq_totale"

```python
# Define the bins and corresponding labels
bins = [0, 5, 10, 15, 20, 30, 40, 50, 100, 200]
labels = ["0-5", "5-10", "10-15", "15-20", "20-30", "30-40", "40-50", "50-100", "100-200"]

# Create the binned column
df['freq_bins'] = pd.cut(df['freq_totale'], bins=bins, labels=labels, right=False)

# Verify the bins
print(df[['freq_totale', 'freq_bins']].head())
```

```python
# Group by bins and calculate the mean reactions
avg_reactions_per_bin = df.groupby('freq_bins')['freq_react'].mean()
avg_reactions_per_bin_df = avg_reactions_per_bin.reset_index()
avg_reactions_per_bin_df.columns = ['Frequency Bin', 'Average Reactions']

# Print the data for reference
print(avg_reactions_per_bin_df)

# Identify the capping value (beginning of plateau or max reaction)
# Find the index of the highest average reaction
max_reaction_index = avg_reactions_per_bin_df['Average Reactions'].idxmax()
capping_bin = avg_reactions_per_bin_df.iloc[max_reaction_index]['Frequency Bin']
capping_reaction = avg_reactions_per_bin_df.iloc[max_reaction_index]['Average Reactions']

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(avg_reactions_per_bin_df['Frequency Bin'], avg_reactions_per_bin_df['Average Reactions'], marker='o', linestyle='-')
plt.title("Average Number of Reactions by Advertisement Bin")
plt.xlabel("Number of Advertisements")
plt.ylabel("Average Reaction")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

# Add a vertical line and annotation for the capping value
plt.axvline(x=capping_bin, color='red', linestyle='--', label=f"Capping Value ({capping_bin})")
plt.annotate(f'Capping Value\n({capping_bin})', 
             xy=(capping_bin, capping_reaction), 
             xytext=(capping_bin, capping_reaction + 0.5),
             # arrowprops=dict(facecolor='red', arrowstyle='->'),
             color='red')

plt.legend()
plt.show()
```

```python
# Filter the data for the bin "30-40"
bin_30_40_data = df[(df['freq_totale'] >= 30) & (df['freq_totale'] < 40)]

# Group by exact number of advertisements within the bin and calculate the average reactions
avg_reactions_30_40 = bin_30_40_data.groupby('freq_totale')['freq_react'].mean()

# Convert to a DataFrame for better readability
avg_reactions_30_40_df = avg_reactions_30_40.reset_index()
avg_reactions_30_40_df.columns = ['Number of Advertisements', 'Average Reaction']

# Find the number of advertisements with the highest average reaction
max_reaction_row = avg_reactions_30_40_df.loc[avg_reactions_30_40_df['Average Reaction'].idxmax()]
capping_value = max_reaction_row['Number of Advertisements']
capping_reaction = max_reaction_row['Average Reaction']

print("Capping Value Found in 30-40 Bin:")
print(max_reaction_row)

# Plot the average reactions for 30-40 range

plt.figure(figsize=(10, 6))
plt.plot(avg_reactions_30_40_df['Number of Advertisements'], avg_reactions_30_40_df['Average Reaction'], marker='o', linestyle='-')
plt.title("Average Reaction for Advertisements in 30-40 Bin")
plt.xlabel("Number of Advertisements")
plt.ylabel("Average Reaction")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add a vertical line for the capping value
plt.axvline(x=capping_value, color='red', linestyle='--', label=f"Capping Value ({int(capping_value)})")
plt.annotate(f'Capping Value\n({int(capping_value)})', 
             xy=(capping_value, capping_reaction), 
             xytext=(capping_value + 0.5, capping_reaction + 0.5),
             # arrowprops=dict(facecolor='red', arrowstyle='->'),
             color='red')

plt.legend()
plt.show()
```

##### Conclusion: The total capping value of advertising is 36 over all customer segments.

## Plotting Weighted Cap Curve

```python
# Define weights
weight_page_view = 1
weight_add_to_cart = 2
weight_order = 3

# Normalize each attribute to the range [0, 1]
df['Product Page View Norm'] = df['Product Page View'] / df['Product Page View'].max()
df['Add to cart Norm'] = df['Add to cart'] / df['Add to cart'].max()
df['Order Norm'] = df['Order'] / df['Order'].max()

# Calculate the weighted reaction using normalized values
df['weighted_reaction'] = (
    df['Product Page View Norm'] * weight_page_view +
    df['Add to cart Norm'] * weight_add_to_cart +
    df['Order Norm'] * weight_order
)
```

```python
# Filter the data for the bin "15-20"
bin_15_20_data = df[(df['freq_prog'] >= 15) & (df['freq_prog'] < 20)]

# Group by exact number of advertisements within the bin and calculate the average weighted reactions
avg_weighted_reactions_15_20 = bin_15_20_data.groupby('freq_prog')['weighted_reaction'].mean()

# Convert to a DataFrame for better readability
avg_weighted_reactions_15_20_df = avg_weighted_reactions_15_20.reset_index()
avg_weighted_reactions_15_20_df.columns = ['Number of Advertisements', 'Average Weighted Reaction']

# Find the number of advertisements with the highest average weighted reaction
max_reaction_row = avg_weighted_reactions_15_20_df.loc[avg_weighted_reactions_15_20_df['Average Weighted Reaction'].idxmax()]
capping_value = max_reaction_row['Number of Advertisements']
capping_reaction = max_reaction_row['Average Weighted Reaction']

print("Capping Value Found in 15-20 Bin:")
print(max_reaction_row)

# Plot the average weighted reactions for the 15-20 range
plt.figure(figsize=(10, 6))
plt.plot(avg_weighted_reactions_15_20_df['Number of Advertisements'], avg_weighted_reactions_15_20_df['Average Weighted Reaction'], marker='o', linestyle='-')
plt.title("Average Weighted Reactions by Programmatic Advertisement Frequency 15-20 Bin")
plt.xlabel("Programmatic Advertisement Frequency")
plt.ylabel("Average Weighted Reaction")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add a vertical line for the capping value
plt.axvline(x=capping_value, color='red', linestyle='--', label=f"Capping Value ({int(capping_value)})")
plt.annotate(f'Capping Value\n({int(capping_value)})', 
             xy=(capping_value, capping_reaction), 
             xytext=(capping_value + 0.5, capping_reaction + 0.02),
             # arrowprops=dict(facecolor='red', arrowstyle='->'),
             color='red')

plt.legend()
plt.show()
```
